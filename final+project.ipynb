```python
import pandas as pd
import numpy as np
import matplotlib as mpl 
import matplotlib.pyplot as plt 
```

```python
# import data 
close_data=pd.read_csv('/Users/longxiaoyi/Desktop/data/close.csv').dropna(axis=1,how='all')
volumn_Data=pd.read_csv('/Users/longxiaoyi/Desktop/data/volumn.csv')
low_Data=pd.read_csv('/Users/longxiaoyi/Desktop/data/lowest.csv')
high_Data=pd.read_csv('/Users/longxiaoyi/Desktop/data/highest.csv')
open_Data=pd.read_csv('/Users/longxiaoyi/Desktop/data/open.csv')
turn_Data=pd.read_csv('/Users/longxiaoyi/Desktop/data/turn.csv')
close_data
```

```python
# delete stocks have no data
left_stock=list(close_data.columns)
volumn_stock=list(volumn_Data.columns)
drop_stock=[]
for i in range(len(volumn_stock)):
    if volumn_stock[i] not in left_stock:
        drop_stock.append(volumn_stock[i])

volumn_data=volumn_Data.drop(drop_stock,axis=1)
low_data=low_Data.drop(drop_stock,axis=1)
high_data=high_Data.drop(drop_stock,axis=1)
open_data=open_Data.drop(drop_stock,axis=1)
turn_data=turn_Data.drop(drop_stock,axis=1)
```

```python
# define many factors and test its efficiency

# 1.define MTM1 :input close_data 
def MTM1(List):
    MTM=np.zeros(int(len(List)/20))
    for i in range(int(len(List)/20)):
        MTM[i]=(List[20*(i+1)-1]-List[i])/List[i]
    return MTM.round(4)

# 2.turnover rate(TR) is already input: input turn_data
def TURN1(List):
    turn=np.zeros(int(len(List)/20))
    for i in range(len(turn)):
        turn[i]=List[20*i:20*(i+1)-1].mean()
    return turn.round(4)
    

#3.turnover rate change(TRC): input turn_data
def TRC1(List):
    TRC=np.zeros(int(len(List)/20))
    for i in range(int(len(List)/20)):
        TRC[i]=List[20*(i+1)-1]-List[i]
    return TRC

#4.average volatility: input close_data
def VAR1(List):
    VAR=np.zeros(int(len(List)/20))
    for i in range(int(len(List)/20)):
        VAR[i]=List[20*i:20*(i+1)-1].var()
    return VAR.round(6)

#5.average volatility change: input close_data
def VARC1(List):
    VARC=np.zeros(int(len(List)/20))
    VARC[0]=0                                               #need change
    for i in range(int(len(List)/20)-1):
        VARC[i+1]=VAR1(List)[i+1]-VAR1(List)[i]
    return VARC.round(4)

#6.RSI indicator :input close_data
def RSI1(List):
    RSI=np.zeros(int(len(List)/20))
    for i in range(int(len(List)/20)):
        rs_up=0
        rs_down=0
        up=0
        down=0
        for j in range(19):
            if List[20*i+1+j]-List[20*i+j]>0:
                rs_up+=abs(List[20*i+1+j]-List[20*i+j])
                up+=1
            if List[20*i+1+j]-List[20*i+j]<0:
                rs_down+=abs(List[20*i+1+j]-List[20*i+j])
                down+=1
        if up!=0 and down!=0:
            RSI[i]=100-100/((rs_up/up)/(rs_down/down)+1)
        if up==0 or down==0:
            RSI[i]=0
    return RSI.round(4)

```

```python
# calculate indicators matrix
indicator_data=pd.DataFrame(columns=close_data.keys(),index=['MTM','turn','TRC','VAR','VARC','RSI'])

for i in range(len(indicator_data.iloc[1,:])):
    indicator_data.iloc[:,i]=[MTM1(close_data.iloc[:,i]),TURN1(turn_data.iloc[:,i]),TRC1(turn_data.iloc[:,i]),VAR1(close_data.iloc[:,i]),VARC1(close_data.iloc[:,i]),RSI1(close_data.iloc[:,i])]
```
